# HEALTH HEART VISUALISATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("Health_heart_experimental.csv")
df.head(5)
```

## LINE CHART

```python
df_line = df.groupby('age')['SysBP'].mean().reset_index()

plt.figure()
plt.plot(df_line['age'], df_line['SysBP'], color='black')

plt.title(" Systolic Blood Pressure ")
plt.xlabel("Age")
plt.ylabel(" Systolic Blood Pressure ")
plt.show()
```

Description: The above line plot shows the systolic blood pressure for each age. The x-axis represents age, while the y-axis represents the systolic blood pressure. The plot helps visualize how blood pressure changes across different age groups in the dataset. From the data, SysBP starts to lower in younger ages (21-30 years) and rises as age increases (60 years) from graph, which means blood pressure rises with age which shows health risks like hypertension in older people.




## BAR PLOT

```python
df['sex'].value_counts().plot(kind='bar', color=['pink','lightblue'])
plt.xlabel(" Sex (0 = Female, 1 = Male)")
plt.ylabel("Count")
plt.title("Sex Distribution")
plt.show()
```

Description: This barplot shows the number of  males and females in the dataset (0 for Female, 1 for Male).The x-axis shows sex and y-axis shows count. It uses bars to compare how many are in each category. In the above data, males have about 3 counts and females have 2. This means the dataset has more men than female.

## PIE CHART

```python
df['indication'].value_counts().plot(kind='pie',autopct='%1.1f%%', colors=['grey','pink'])
plt.title("Heart Disease Indication")
plt.ylabel("")
plt.show()
```

Description: This pie chart displays how many people in the dataset have heart disease (1 = Yes) or not (0 = No).About 47.3% have heart disease and 52.7% do not. So, most are healthy but some have heart risks like high blood pressure and BMI. This provides a clear overview of the health condition distribution.

## BOX PLOT

```python
plt.figure()
plt.boxplot(df['BMI'])
plt.ylabel("BMI")
plt.title("BMI Distribution")
plt.show()
```

Description: The above box plot shows the BMI (Body Mass Index) values from the dataset.It helps identify people with very high or very low BMI. The box shows BMI mostly between about 20 and 32, with the middle line around 25-30. Some high values above 32 means overweight people and low values i.e below 20 are underweight. 



## HISTOGRAM

```python
plt.hist(df['HR'], bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Heart Rate")
plt.ylabel("Frequency")
plt.title("Heart Rate Distribution")
plt.show()
```

Description: The histogram shows the frequency of different heart rate values. It uses 5 bars.The x-axis shows heart rate and y-axis shows frequency of different heart rates. Most people have HR between 100-130 bpm, and the highest is around 120-130. High values have health risks like high BP.

## SCATTER PLOT

```python
sns.scatterplot(x='heightCm', y='weightKg',data=df,)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Weight vs Height ")
plt.show()
```

Description:The scatter plot shows how weight increases with height.The x-axis shows height in cm and y-axis shows weight in kg. Most dots cluster around 150-180 cm height and 50-90 kg weight.High weight at average height might risk to high BMI. This plot helps to understand the overall distribution of body sizes in the dataset.
