### Matrix Operations

```python
A = np.array([[2, -1],[0, 3]])
print(A.shape)
B = np.array([
    [1, 2],
    [-3, 4]
])
print(B.shape)
```

```python
print("Addition of two matrices A and B is : ")
print(A + B)
```

```python
print("The Dot Product of A and B is: ")
print(A @ B)
```

```python
print(" The expression \"3*A - 2*B\"")
print(3 * A - 2 * B)
```

##  Determinants and Inverses  

```python
import numpy.linalg as LA
```

```python
print("Determinants of A is:")
A = np.array([
    [2, -1],
    [0, 3]
])
print(A)
LA.det(A)
```

Determinant of A is 6 . Hence, it is a non-singular matrix and it is invertible.


# A^-1

```python
print("Inverse of the matrix A :")
print(A)
print( LA.inv(A))
```

```python
X = np.array([
    [4, 1.5, 100],
    [6, 2.0, 150]
])

y = np.array([20, 30])
```

```python
beta = np.linalg.pinv(X) @ y
```

```python
car3 = np.array([8, 3.0, 200])
price = car3 @ beta 
print(" Car 3:")
print(car3)
print("Price of the car 3 is ",price)
```
